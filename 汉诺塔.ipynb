# 汉诺塔
    描述：有A,B,C三个塔，把A上面的盘子全部移动到C盘上
# 汉诺塔规则
    1.每次只能挪动一个盘子
    2.大盘子必须在小盘子上面

def hano(n,a,b,c):
    if(n==1):
        print(a,"--->",c)
        return None
    
    if(n==2):
        print(a,"--->",b)
        print(a,"--->",c)
        print(b,"--->",c)
        return None
    
    hano(n-1,a,c,b)
    print(a,"--->",c)
    hano(n-1,b,a,c)

```python
a = "A"
b = "B"
c = "C"
n = 2
hano(n,a,b,c)
```

```python
n = 1
hano(n,a,b,c)
```

```python
n = 3
hano(n,a,b,c)
```

```python
n = 10
hano(n,a,b,c)
```
